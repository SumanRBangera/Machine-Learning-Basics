```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Orders1.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['State'].hist(bins=30)
plt.show()
```

```python
data['City'].hist(bins=30)
plt.show()
```

```python
# Bar chart
data['Category'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.scatter(data['Order ID'], data['State'])
plt.xlabel('Order')
plt.ylabel('State')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='Order ID', y='State', data=data)
plt.show()
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
